## The Ames Housing Dataset

The [Ames Housing dataset](http://jse.amstat.org/v19n3/decock.pdf) is a newer (2011) replacement for the classic Boston Housing dataset. Each record represents a residential property sale in Ames, Iowa. It contains many different potential predictors and the target variable is `SalePrice`.

````python
import pandas as pd
ames = pd.read_csv("ames.csv", index_col=0)
ames
````

````python
ames.describe()
````

We will focus specifically on a subset of the overall dataset. These features are:

```
LotArea: Lot size in square feet

1stFlrSF: First Floor square feet

GrLivArea: Above grade (ground) living area square feet
```

````python
ames_subset = ames[['LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].copy()
ames_subset
````

## Step 1: Visualize Relationships Between Features and Target

For each feature in the subset, create a scatter plot that shows the feature on the x-axis and `SalePrice` on the y-axis.

````python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# fearture 1: GrLivArea
# defining x and y
x = ames_subset['GrLivArea']
y = ames_subset['SalePrice']

# Add constant to predictor (for intercept)
x = sm.add_constant(x)

# Fit OLS model
model = sm.OLS(y, x).fit()

# Print model summary
print(model.summary())

# Plot regression line
plt.figure(figsize=(8, 5))
sns.scatterplot(x=ames_subset['GrLivArea'], y=ames_subset['SalePrice'], alpha=0.6)
plt.plot(ames_subset['GrLivArea'], model.predict(x), color='red')
plt.title('GrLivArea vs SalePrice with Regression Line')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.show()
````

````python
# fearture 2: LotArea
# defining x and y
x = ames_subset['LotArea']
y = ames_subset['SalePrice']

# Add constant to predictor (for intercept)
x = sm.add_constant(x)

# Fit OLS model
model = sm.OLS(y, x).fit()

# Print model summary
print(model.summary())

# Plot regression line
plt.figure(figsize=(8, 5))
sns.scatterplot(x=ames_subset['LotArea'], y=ames_subset['SalePrice'], alpha=0.6)
plt.plot(ames_subset['LotArea'], model.predict(x), color='red')
plt.title('LotArea vs SalePrice with Regression Line')
plt.xlabel('LotArea')
plt.ylabel('SalePrice')
plt.show()
````

````python
# fearture 3: 1stFlrSF
# defining x and y
x = ames_subset['1stFlrSF']
y = ames_subset['SalePrice']

# Add constant to predictor (for intercept)
x = sm.add_constant(x)

# Fit OLS model
model = sm.OLS(y, x).fit()

# Print model summary
print(model.summary())

# Plot regression line
plt.figure(figsize=(8, 5))
sns.scatterplot(x=ames_subset['1stFlrSF'], y=ames_subset['SalePrice'], alpha=0.6)
plt.plot(ames_subset['1stFlrSF'], model.predict(x), color='red')
plt.title('1stFlrSF vs SalePrice with Regression Line')
plt.xlabel('1stFlrSF')
plt.ylabel('SalePrice')
plt.show()
````

* `GrLivArea` -> Strong Predictor and a good candidate for Simple linear regression
 
* `1stFlrSF` -> Moderate predictor and possibly can be used, but it may not capture the `SalePrice` as well as `GrLivArea` alone

* `LotArea` - > Weak predictor and is not a strong candidate for simple regression on its own


## Step 2: Build a Simple Linear Regression Model

Set the dependent variable (`y`) to be the `SalePrice`, then choose one of the features shown in the subset above to be the baseline independent variable (`X`).

Build a linear regression using StatsModels, describe the overall model performance, and interpret its coefficients.

````python
# Define the dependent variable
y = ames_subset['SalePrice']

# Define the independent variable (baseline feature)
x = ames_subset['GrLivArea']

# Add a constant to the independent variable (intercept term)
X = sm.add_constant(x)
````

````python
import statsmodels.api as sm

# Fit the OLS model
model = sm.OLS(y, X).fit()

# Display the summary of the regression results
print(model.summary())
````

R-squared = 0.502
About 50.2% of the variability in SalePrice is explained by GrLivArea. This is a moderate fit — not bad for a single-variable model, but it leaves room for improvement with more predictors.

F-statistic = 1471.0 and p-value (Prob F) ≈ 0.000
The model overall is statistically significant, meaning GrLivArea does help predict SalePrice.

## Step 3: Build a Multiple Linear Regression Model

For this model, use **all of** the features in `ames_subset`.

````python
# Define X and y for multiple linear regression
X = ames_subset[['LotArea', '1stFlrSF', 'GrLivArea']]
y = ames_subset['SalePrice']

# Add a constant term to X for the intercept
X = sm.add_constant(X)

# Fit the model
model_multiple = sm.OLS(y, X).fit()

# Display results
print(model_multiple.summary())
````

Model Fit:
R-squared = 0.565 vs. 0.502 in the simple model.

This means the multiple regression model explains 56.5% of the variation in SalePrice, an improvement over the previous model's 50.2%.

So yes, the model is a better fit than before.

Coefficients:

GrLivArea remains the most significant predictor, with a coefficient of 80.6 and a very low p-value (<0.001), confirming its strong positive influence on price.

1stFlrSF is also highly significant, with a positive effect.

LotArea has a borderline p-value (0.051), suggesting it's not strongly significant, and its coefficient is much smaller. It may contribute little additional explanatory power.

## Step 4: Create Partial Regression Plots for Features

Using your model from Step 3, visualize each of the features using partial regression plots.

````python
# Define X and y
X = ames_subset[['LotArea', '1stFlrSF', 'GrLivArea']]
y = ames_subset['SalePrice']

# Add a constant term to X
X = sm.add_constant(X)

# Fit the multiple linear regression model
model = sm.OLS(y, X).fit()

# Create partial regression plots
fig = plt.figure(figsize=(12, 8))
sm.graphics.plot_partregress_grid(model, fig=fig)
plt.tight_layout()
plt.show()
````

### Partial Regression Plot Summary

- **GrLivArea**: Still shows a strong linear relationship with `SalePrice`, confirming it’s a key predictor even after controlling for other variables.
- **1stFlrSF**: Positive relationship remains, though weaker than `GrLivArea`, indicating it still contributes meaningfully.
- **LotArea**: Shows the weakest relationship; contribution to predicting `SalePrice` is minor after accounting for other features.
